```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def p(x):
    return 1 + x*x
def q(x):
    return 4*x**3 - 6*x**2 + 2.25*x
def f(x):
    return -8*math.exp(-x**2)*(1.25*x**2 - 2.75*x - 2)


[a, b] = [0, 3]
h = 1

```

```python
k = 3
```

```python
def P(i):
    return p(a + i*h)
def Q(i):
    return q(a + i*h)
def F(i):
    return f(a + i*h)
```

```python
n = int((b-a)/h)
print('n = ', n)
M = np.eye(n-1)

M[0, 1] = P(3/2)
M[0, 0] = -(M[0, 1] + h*h*Q(1) + P(1/2))
M[n-2, n-3] = P(n-1-1/2)
M[n-2, n-2] = -(M[n-2, n-2] + h*h*Q(n-1) + P(n-1+1/2))

for i in range(1, n-2):
    M[i, i-1] = P(i+1-1/2)
    M[i, i+1] = P(i+1+1/2)
    M[i, i] = -(M[i, i+1] + M[i, i-1] + h*h*Q(i+1))

D = np.eye(n-1)
for i in range(n-1):
    D[i, i] = -1*h*h*F(i+1)

print("M =", M)
print("D = ", D)
print("Giá trị riêng và hàm riêng cần tìm là giá trị riêng và vector riêng của D^{-1}M")
D_inverseM = np.dot(np.linalg.inv(D), M)

eigenvalues, eigenvectors = np.linalg.eig(D_inverseM)

X = np.arange(a, b+h, h)


SoLuongTriRieng = len(eigenvalues)
for i in range(SoLuongTriRieng):
    eigenfunction = np.array(eigenvectors[:, i])
    eigenfunction = np.insert(eigenfunction, 0, 0)
    eigenfunction = np.append(eigenfunction, 0)
    print("Trị riêng: ", eigenvalues[i])
    print("Hàm riêng tương ứng:")
    plt.plot(X, eigenfunction, label = 'siuuuuu') # Vẽ y_1 theo t
    plt.xlabel('X')
    plt.ylabel('Y') 
    plt.legend()
    plt.title('😗~Siuuuuu ')
    plt.show()


```
